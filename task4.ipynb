## Prewitt filters

**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img=plt.imread("data/astronaut.png")
plt.imshow(img)
```

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] <1 or p[0]>=(img_in.shape[0]-1) or p[1] <1 or p[1]>=(img_in.shape[1]-1):
            result[p]=0
        else:
            for q in numpy.ndindex(3, 3):
                if q[1]==0:
                    neighborhood_sum=neighborhood_sum-img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                if q[1]==2:
                    neighborhood_sum=neighborhood_sum+img_in[q[0]+p[0]-1,q[1]+p[1]-1]
            result[p]=neighborhood_sum/6
    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] <1 or p[0]>=(img_in.shape[0]-1) or p[1] <1 or p[1]>=(img_in.shape[1]-1):
            result[p]=0
        else:
            for q in numpy.ndindex(3, 3):
                if q[0]==0:
                    neighborhood_sum=neighborhood_sum-img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                if q[0]==2:
                    neighborhood_sum=neighborhood_sum+img_in[q[0]+p[0]-1,q[1]+p[1]-1]
            result[p]=neighborhood_sum/6
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
img_own_prewitt_h=prewitt_h(img)
plt.imshow(img_own_prewitt_h)
```

```python
img_own_prewitt_v=prewitt_v(img)
plt.imshow(img_own_prewitt_v)
```

**<a id="#">Task 4.1.4.</a>** Compare your results with the correct result image `data/astronaut_prewitt_h.tiff` and `data/astronaut_prewitt_v.tiff`:

```python
import skimage.io as ski

img_their_prewitt_h=ski.imread("data/astronaut_prewitt_h.tiff")
plt.imshow(img_their_prewitt_h)
assert numpy.allclose(img_own_prewitt_h, img_their_prewitt_h, atol=1/255)

img_their_prewitt_v=ski.imread("data/astronaut_prewitt_v.tiff")
plt.imshow(img_their_prewitt_v)
assert numpy.allclose(img_own_prewitt_v, img_their_prewitt_v, atol=1/255)
```

## Sobel filters (bonus)

**<a id="#">Task 4.3.1.</a>** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical 3&times;3 Sobel filtering, respectively:

```python
def sobel_h(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] <1 or p[0]>=(img_in.shape[0]-1) or p[1] <1 or p[1]>=(img_in.shape[1]-1):
            result[p]=0
        else:
            for q in numpy.ndindex(3, 3):
                if q[1]==0:
                    neighborhood_sum=neighborhood_sum-img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                if q[1]==2:
                    neighborhood_sum=neighborhood_sum+img_in[q[0]+p[0]-1,q[1]+p[1]-1]
            result[p]=neighborhood_sum/8
    return result
    
def sobel_v(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] <1 or p[0]>=(img_in.shape[0]-1) or p[1] <1 or p[1]>=(img_in.shape[1]-1):
            result[p]=0
        else:
            for q in numpy.ndindex(3, 3):
                if q[0]==0:
                    if q[1]==1:
                        neighborhood_sum=neighborhood_sum-2*img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                    else:
                        neighborhood_sum=neighborhood_sum-img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                if q[0]==2:
                    if q[1]==1:
                        neighborhood_sum=neighborhood_sum+2*img_in[q[0]+p[0]-1,q[1]+p[1]-1]
                    else:
                        neighborhood_sum=neighborhood_sum+img_in[q[0]+p[0]-1,q[1]+p[1]-1]
            result[p]=neighborhood_sum/8
    return result
```

**<a id="#">Task 4.3.2.</a>** Test your above solution using the functions `sobel_h` and `sobel_v` for the previously loaded image and show the result:

```python
img_own_sobel_h=sobel_h(img)
plt.imshow(img_own_sobel_h)
```

```python
img_own_sobel_v=sobel_v(img)
plt.imshow(img_own_sobel_v)
```

```python
img_their_sobel_v=ski.imread("data/astronaut_sobel_v.tiff")
plt.imshow(img_their_sobel_v)
assert numpy.allclose(img_own_sobel_v, img_their_sobel_v, atol=1/255)
```
